### Building Model and Feature Engineering

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
le_Gender = LabelEncoder()
```

```python
df['Income_Category_Labelled'] = le_Gender.fit_transform(df['Income_Category'])
df['Education_Level_Labelled'] = le_Gender.fit_transform(df['Education_Level'])
df['Card_Category_Labelled'] = le_Gender.fit_transform(df['Card_Category'])
```

```python
df = df.drop([ 'Education_Level',  'Income_Category', 'Card_Category','CLIENTNUM'], axis = 1)
```

```python
df['Attrition_Flag'] = df['Attrition_Flag'].map({'Existing Customer' : 1,'Attrited Customer':0})
```

```python
new_df = pd.get_dummies(df, columns=['Gender','Marital_Status'], drop_first=True, dtype = 'int')
```

```python
new_df.dtypes
```

```python
new_df = new_df.drop('util_new', axis = 1)
```

```python
new_df.head()
```

```python
train = new_df.drop('Attrition_Flag', axis = 1)
test = new_df['Attrition_Flag']
```

```python
x_train, x_test, y_train, y_test = train_test_split(train, test, test_size = 0.19, random_state = 20)
```

```python
# Linear Regression
lr_model = LogisticRegression()
lr_model.fit(x_train, y_train)
lr_predictions = lr_model.predict(x_test)
lr_accuracy = accuracy_score(y_test, lr_predictions)

# Naive Bayes Classifier
nb_model = GaussianNB()
nb_model.fit(x_train, y_train)
nb_predictions = nb_model.predict(x_test)
nb_accuracy = accuracy_score(y_test, nb_predictions)
```

### Model Validation (Accuracy, Classification report)

```python
# Print accuracy for each model
print("Linear Regression: ", lr_accuracy)
print("Naive Bayes Accuracy:", nb_accuracy)

# Generate classification reports for each model
print("\nLinear Regression Report:")
print(classification_report(y_test, lr_predictions))

print("\nNaive Bayes Classification Report:")
print(classification_report(y_test, nb_predictions))
```
